```python
import sys
import os

sys.path.append('../')

from toloof import *

def plot_enmap(tmpmap,plot_vmin = None,plot_vmax = None,save_fig_name = None,cbar_label=None,plot_title=None,xlims=None,ylims=None):
	corners_tmp = np.rad2deg(enmap.corners(tmpmap.shape,tmpmap.wcs))
	imextent_tmp = [corners_tmp[0,1],corners_tmp[1,1],corners_tmp[0,0],corners_tmp[1,0]]
	fig=plt.figure(figsize=(8,8))
	im = plt.imshow(tmpmap,extent=imextent_tmp,origin='lower',vmin=plot_vmin,vmax=plot_vmax)
	plt.xticks(fontsize=10)
	plt.yticks(fontsize=10)
	plt.xlabel('Right Ascension (deg)')
	plt.ylabel('Declination (deg)')
	plt.title(plot_title)
	plt.xlim(xlims)
	plt.ylim(ylims)
	cbar_ax = fig.add_axes([0.91, 0.15, 0.01, 0.7])
	cbar = fig.colorbar(im, cax=cbar_ax)
	cbar.set_label(label=cbar_label,rotation=270,labelpad=15,fontsize=14)    
	if save_fig_name is not None:
		plt.savefig(save_fig_name,facecolor='white',transparent=False,bbox_inches='tight')
	
	plt.show()
```

```python
paths2file = np.array(['test_data/134272/raw/toltec_commissioning_a2000_pointing_134272_citlali.fits','test_data/134273/raw/toltec_commissioning_a2000_pointing_134273_citlali.fits','test_data/134274/raw/toltec_commissioning_a2000_pointing_134274_citlali.fits'])
#paths2file = np.array(['test_data/134272/raw/toltec_commissioning_a1100_pointing_134272_citlali.fits','test_data/134273/raw/toltec_commissioning_a1100_pointing_134273_citlali.fits','test_data/134274/raw/toltec_commissioning_a1100_pointing_134274_citlali.fits'])

tmp = Fraunhofer_Beamfit(paths2file,2E-3,padpixels = 50)
#tmp = Fraunhofer_Beamfit_new(paths2file,1.1E-3)
```

```python
tmp.initialize_model(aperture_plane_resolution = 1.0,bandstr='band_220',interpwavelengths=300E-3/np.linspace(195,242,5),
                    aperture_fwhm = 43.,edge_taper_diameter=43.)

```

```python
# make some bounds

src_amp_lower = 0.6*np.amax(tmp.trunc_maps['map1'])
src_amp_upper = 1.4*np.amax(tmp.trunc_maps['map1'])

primary_foc_lower = 17.
primary_foc_upper = 18.

m2x_lower = -10E-3
m2x_upper = 10E-3

m2y_lower = -10E-3
m2y_upper = 10E-3

m2z_lower = -10E-3
m2z_upper = 10E-3

TILT_Y_lower = -500.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
TILT_Y_upper = 500.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

TILT_X_lower = -500.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
TILT_X_upper = 500.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

AST_O_lower = -300.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
AST_O_upper = 300.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

AST_V_lower = -300.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
AST_V_upper = 300.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

TRE_V_lower = -200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
TRE_V_upper = 200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

COMA_V_lower = -200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
COMA_V_upper = 200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

COMA_H_lower = -200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
COMA_H_upper = 200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

TRE_O_lower = -200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
TRE_O_upper = 200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

QUAD_O_lower = -200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
QUAD_O_upper = 200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

AST2_O_lower = -200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
AST2_O_upper = 200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

SPH_lower = -200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
SPH_upper = 200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

AST2_V_lower = -200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
AST2_V_upper = 200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

QUAD_V_lower = -200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength
QUAD_V_upper = 200.*(1E-6)*(2.*np.pi)*np.sqrt(2)/tmp.wavelength

total_bounds = ((src_amp_lower,src_amp_upper),(primary_foc_lower,primary_foc_upper),(m2x_lower,m2x_upper),
                (m2y_lower,m2y_upper),(m2z_lower,m2z_upper),(TILT_Y_lower,TILT_Y_upper),(TILT_X_lower,TILT_X_upper),
                (AST_O_lower,AST_O_upper),(AST_V_lower,AST_V_upper),(TRE_V_lower,TRE_V_upper),(COMA_V_lower,COMA_V_upper),
                (COMA_H_lower,COMA_H_upper),(TRE_O_lower,TRE_O_upper),(QUAD_O_lower,QUAD_O_upper),(AST2_O_lower,AST2_O_upper),
                (SPH_lower,SPH_upper),(AST2_V_lower,AST2_V_upper),(QUAD_V_lower,QUAD_V_upper))

lower_bound_array = np.array([src_amp_lower,primary_foc_lower,m2x_lower,m2y_lower,m2z_lower,
                              TILT_Y_lower,TILT_X_lower,AST_O_lower,AST_V_lower,TRE_V_lower,
                              COMA_V_lower,COMA_H_lower,TRE_O_lower,QUAD_O_lower,AST2_O_lower,
                              SPH_lower,AST2_V_lower,QUAD_V_lower])

upper_bound_array = np.array([src_amp_upper,primary_foc_upper,m2x_upper,m2y_upper,m2z_upper,
                              TILT_Y_upper,TILT_X_upper,AST_O_upper,AST_V_upper,TRE_V_upper,
                              COMA_V_upper,COMA_H_upper,TRE_O_upper,QUAD_O_upper,AST2_O_upper,
                              SPH_upper,AST2_V_upper,QUAD_V_upper])


tmp.fit_beam(c_guess=None,boundvals = total_bounds,fit_achromatic_beam=False)
```
